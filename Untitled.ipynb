```python
type Period = Literal['1d', '5d', '1mo', '3mo', '6mo', '1y', '2y', '5y', '10y', 'ytd', 'max']
type Interval = Literal['1m', '2m', '5m', '15m', '30m', '60m', '90m', '1h', '4h', '1d', '5d', '1wk', '1mo', '3mo']

def set_timezone(dt_index: pd.DatetimeIndex, target_tz: str = "Etc/GMT-1") -> pd.DatetimeIndex:
    if dt_index.tz and dt_index.tz.tzname != target_tz:
        dt_index = dt_index.tz_convert(target_tz)

    return dt_index


def fetch_data_yfinance(symbols: list[str], interval: Interval, period: Period, prepost: bool = False) -> dict[str, pd.DataFrame]:
    data = yf.download(
        symbols, 
        period=period,
        interval=interval,
        group_by='ticker',
        auto_adjust=True,
        threads=True,  # Parallel
        prepost=prepost # Premarket
    )

    result = {}

    if data is None or data.empty:
        return result

    tickers = data.columns.get_level_values('Ticker')
    valid_symbols = [symbol for symbol in symbols if symbol in tickers]
    
    for symbol in valid_symbols:
        df = data[symbol].copy()

        df.index = set_timezone(pd.to_datetime(df.index), target_tz="Etc/GMT-1")

        df['Volatility'] = ((df['High'] - df['Low']) / df['Close'] * 100)
        df['Change'] = ((df['Close'] - df['Open']) / df['Open'] * 100)

        df = df.astype({
            'Volatility': float,
            'Change': float,
            'Volume': int
        })
            
        result[symbol] = df
    
    return result
```

# Utilities

```python
def normalize(series):
    # convert to float and replace inf/-inf with NaN
    series = series.astype(float).replace([np.inf, -np.inf], np.nan)
    
    s = series.dropna()
    
    if s.empty or s.max() == s.min():  # avoid division by zero
        return pd.Series(0, index=series.index)
    
    return (series - s.min()) / (s.max() - s.min())


def normalize_signed(series):
    # Convert to float and replace inf/-inf with NaN
    series = series.astype(float).replace([np.inf, -np.inf], np.nan)
    
    s = series.dropna()
    
    if s.empty or s.std() == 0:  # avoid division by zero
        return pd.Series(0, index=series.index)
    
    max_abs = max(abs(s.min()), abs(s.max()))
    return series / max_abs  # scales to [-1, 1]


def true_range(df):
    high_low = df['High'] - df['Low']
    high_close = (df['High'] - df['Close'].shift()).abs()
    low_close = (df['Low'] - df['Close'].shift()).abs()
    
    return pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)

def bollinger_bands(series: pd.Series, length: int = 20, *, num_stds: tuple[float, ...] = (2, 0, -2), prefix: str = '') -> pd.DataFrame:
    # Ref: https://stackoverflow.com/a/74283044/
    rolling = series.rolling(length)
    bband0 = rolling.mean()
    bband_std = rolling.std(ddof=0)
    return tuple((bband0 + (bband_std * num_std)) for num_std in num_stds)

def calculate_rsi_ema(series, period=14):
    delta = series.diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    
    avg_gain = gain.ewm(alpha=1/period, min_periods=period).mean()
    avg_loss = loss.ewm(alpha=1/period, min_periods=period).mean()
    
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    
    return rsi
```

```python
#pdf = df.iloc[120:-120].copy()
pdf = df.copy()
pdf.index = pdf.index.astype(str)

def datecut(df, to=None, fr=None):
    if to:
        mask = df.index.str.startswith(to)
        if mask.any():
            first_idx = df.index[mask][0]
            df = df.loc[:first_idx]
    if fr:
        mask = mask = df.index.str.startswith(fr)
        if mask.any():
            start_idx = df.index[mask][0]
            df = df.loc[start_idx:]
    return df

#pdf = datecut(pdf, fr='2025-11-03')

std_win = 8

pdf['Volume_log'] = np.log2(pdf['Volume'])
pdf['Volume_n'] = normalize(pdf['Volume'])

# EWM seems to fit more snugly to the original volume than rolling
pdf['Volume_ewm'] = pdf[['Volume']].ewm(span=std_win, adjust=False).mean()
pdf['Volume_roll'] = pdf[['Volume']].rolling(window=std_win).mean()
pdf['Volume_ratio'] = pdf['Volume'] / pdf['Volume_ewm'].mean()
pdf['Volume_ratio_pct_change'] = pdf['Volume_ratio'].pct_change()

pdf['Volume_pct_change'] = pdf['Volume'].pct_change().fillna(0)
pdf['Volume_pct_change_n'] = normalize_signed(pdf['Volume_pct_change'])
pdf['Volume_diff'] = pdf['Volume'].diff()

pdf['Close_n'] = normalize(pdf['Close'])
pdf['Close_pct_change'] = pdf['Close'].pct_change().fillna(0)
pdf['Close_pct_change_n'] = normalize_signed(pdf['Close_pct_change'])
pdf['Close_roll_pct_change'] = pdf['Close'].rolling(window=std_win).mean().pct_change()

pdf['Gain'] = pdf['Close'] - pdf['Open']
pdf['Gain_diff'] = pdf['Gain'].diff()
pdf['Resistance'] = pdf['High'].max()
pdf['Support'] = pdf['Low'].min()

pdf['Range'] = pdf['High'] - pdf['Low']
pdf['TR'] = true_range(pdf)
pdf['ATR'] = pdf['TR'].rolling(window=std_win*2).mean()
pdf['Gain_close_ratio'] = pdf['Gain'] / pdf['Close']

# Theory from YT: if candle size > 2*ATR we have potential
# breakout from oscillation between support and resistance
pdf['Break'] = (pdf['Gain'] > 1.6 * pdf['ATR'])

(pdf['Boll_h'], pdf['Boll_m'], pdf['Boll_l']) = bollinger_bands(pdf['Close'], length=std_win*2)
pdf['Boll_w'] = pdf['Boll_h'] - pdf['Boll_l']


pdf['Volume_sentiment'] = np.where(
    pdf['High'] != pdf['Low'],
    (( (pdf['Close'] - pdf['Low']) - (pdf['High'] - pdf['Close']) ) / (pdf['High'] - pdf['Low'])) * pdf['Volume'],
    0
)

pdf['Volume_sentiment_pos'] = (pdf['Volume_sentiment'] > 0)

pdf['Volume_momentum_norm'] = (pdf['Volume'] - pdf['Volume'].rolling(std_win).mean()) / pdf['Volume'].rolling(std_win).std()
pdf['Volume_momentum_norm_n'] = normalize(pdf['Volume_momentum_norm'])

pdf['RSI'] = calculate_rsi_ema(pdf['Close'], std_win*2)

# ----------------------------------------- Vibed metrics ---------------------------------------------

# 1. VOLUME BREAKOUT - massive volume spike
pdf['Volume_change_spike'] = (pdf['Volume_pct_change'] > 5.0)  # 500% volume increase
pdf['Volume_change_spike_near'] = ((pdf['Volume_change_spike']).astype(int).rolling(window=int(std_win)).sum() > 0)

# 2. VOLUME SENTIMENT SURGE - strong buying pressure
rolling_sentiment = pdf['Volume_sentiment'].rolling(std_win*2)
pdf['Bullish_sentiment'] = (pdf['Volume_sentiment'] > (rolling_sentiment.mean() + 2 * rolling_sentiment.std()))

# 3. BOLLINGER BAND SQUEEZE THEN EXPANSION
pdf['Boll_squeeze'] = (pdf['Boll_w'] < pdf['Boll_w'].rolling(std_win*2).quantile(0.2))  # Tight bands
pdf['Boll_squeeze_near'] = ((pdf['Boll_squeeze']).astype(int).rolling(window=int(std_win)).sum() > 0)
pdf['Boll_squeeze_w'] = (pdf['Boll_squeeze']) * pdf['Boll_w']
pdf['Boll_breakout_h'] = (pdf['Close'] > pdf['Boll_h'])
pdf['Boll_breakout_l'] = (pdf['Close'] < pdf['Boll_l'])

# 4. MOMENTUM SURGE - volume momentum spike
pdf['Momentum_spike'] = (pdf['Volume_momentum_norm'] > 2.0)  # 2 std devs above mean

# 5. PRICE + VOLUME CONFIRMATION
pdf['Price_vol_breakout'] = (pdf['Close_roll_pct_change'] > 0.1) & (pdf['Volume_ratio'] > 3)

# 6. SUSTAINED GAIN - multiple green candles
pdf['Sustained_gain'] = (pdf['Gain'] > 0) & (pdf['Gain'].rolling(int(std_win/2)).apply(lambda x: (x > 0).sum()) >= int(std_win/2)*0.75)

# 7. VOLATILITY EXPANSION - ATR spike
pdf['Volatility_spike'] = (pdf['ATR'] > pdf['ATR'].rolling(std_win).mean() * 1.5)

pdf['RSI_pos'] = (pdf['RSI'] >= 60)
pdf['RSI_neg'] = (pdf['RSI'] <= 40)

pdf['Medium_signal'] = (
    (pdf['Boll_breakout_h']) &
    (pdf['Volume_change_spike_near'] | pdf['Volatility_spike']) &
    (pdf['RSI_pos']) &
    (pdf['Gain_diff'] > 0) 
    #(pdf['Boll_squeeze_near'])
)

pdf['Strong_signal'] = (
    (pdf['Medium_signal']) &
    (pdf['Price_vol_breakout']) & 
    (pdf['Break']) & 
    (pdf['Sustained_gain']) &
    (pdf['Volume_sentiment'] > 0) &
    (pdf['Bullish_sentiment'])
)

# --------------------------------------- PLOT ------------------------------------------------

fig, axes = plt.subplots(6,2, figsize=(10, 16), sharex=True)
axes = axes.flatten()

def scatter_dot(ax, column, y, color=None):
    signal_positions = pdf.index[pdf[column].astype(bool)]
    signal_positions = [pdf.index.get_loc(dt) for dt in signal_positions]
    ax.scatter(signal_positions, [y]*len(signal_positions), color=color, s=8, zorder=6, label=column)

ax = axes[0]

midy = max(pdf['Boll_h'].max(), pdf['Close'].max()) / 2
scatter_dot(ax, column='Strong_signal', y=midy*2-0.1, color='black')
scatter_dot(ax, column='Medium_signal', y=midy*2-0.2, color='grey')
scatter_dot(ax, column='Boll_breakout_h', y=0.1*midy, color='green')
scatter_dot(ax, column='Boll_breakout_l', y=0.1*midy, color='red')
scatter_dot(ax, column='Volume_change_spike', y=0.2*midy, color='lime')
scatter_dot(ax, column='Sustained_gain', y=0.3*midy, color='pink')
scatter_dot(ax, column='Momentum_spike', y=0.4*midy, color='blue')
scatter_dot(ax, column='Bullish_sentiment', y=0.5*midy, color='purple')
scatter_dot(ax, column='Volatility_spike', y=0.6*midy, color='cyan')
scatter_dot(ax, column='RSI_pos', y=0.7*midy, color='indigo')


pdf.plot(y='Close', ax=ax)
pdf.plot(y='Boll_h', ax=ax, alpha=0.6, color='skyblue')
pdf.plot(y='Boll_l', ax=ax, alpha=0.6, color='skyblue')
pdf.plot(y=['Boll_w'], kind='bar', ax=ax, alpha=0.1, width=1, color='slategray', title='Close & Bollinger bands (2)')
pdf.plot(y=['Boll_squeeze_w'], kind='bar', ax=ax, alpha=0.2, width=1, color='slategray')

ax = axes[1]

pdf.plot(y=['Volume_n'], ax=ax, kind='bar', alpha=0.4, width=1)
pdf.plot(y='Close_n', ax=ax, kind='line')
ax.set_title('Normalized Close & Volume')

ax = axes[2]
break_points = pdf.index[pdf['Break'].astype(bool)]
break_positions = [pdf.index.get_loc(dt) for dt in break_points]
ax.scatter(break_positions, [0]*len(break_positions), color='blue', s=10, zorder=5, label='Break')
pdf.plot(y=['ATR'], ax=ax, kind='bar', alpha=0.5, width=1, color='skyblue')
pdf.plot(y=['Gain'], ax=ax, kind='bar', alpha=0.5, width=1, color='orange', title='ATR, Gain & Break')

ax = axes[3]
pdf.plot(y=['Volume_diff', 'Volume_sentiment'],ax=ax,kind='bar',title='Volume diff & sentiment', width=1)


ax = axes[4]
pdf.plot(y='RSI',ax=ax,kind='line',title='RSI', color='indigo')
ax.axhline(y=70, color='red', linestyle='--', linewidth=1)
ax.axhline(y=30, color='green', linestyle='--', linewidth=1) 
ax.set_ylim(0, 100) 

ax = axes[5]
pdf.plot(y=['Volume_pct_change'], ax=ax, kind='bar', title='Volume change & ratio', width=1, color=['teal'])
pdf.plot(y=['Volume_ratio'], ax=ax, title='Volume change & ratio', color=['gold'])

pdf.plot(y='Gain_close_ratio',ax=axes[6],kind='bar',title='Gain close ratio')

pdf.plot(y='Volume_momentum_norm',ax=axes[7],kind='bar',title='Volume_momentum_norm')

pdf.plot(y='Volume',ax=axes[8],kind='bar',title='Volume', width=1, color='orange', logy=True)

ax = axes[9]
pdf.plot(y=['Close_roll_pct_change'],ax=ax,kind='bar',title='Close (rolling) pct_change',width=1, color='lightsteelblue')


for ax in axes:
    xticks = ax.get_xticks()
    if len(xticks) > 25:  # only reduce if too many ticks
        ax.set_xticks(xticks[::2])

    labels = []
    for i in ax.get_xticks():
        if 0 <= int(i) < len(pdf):
            # Parse the string index back to datetime for formatting
            dt = pd.to_datetime(pdf.index[int(i)])
            labels.append(dt.strftime('%b %-d %H:%M'))  # Nov 4 10:00
        else:
            labels.append('')
    ax.set_xticklabels(labels, rotation=45, ha='right')
    
    handles, labels = ax.get_legend_handles_labels()
    if handles:
        ax.legend(handles, labels, fontsize=8)
    #ax.legend().set_visible(False)
    ax.tick_params(axis='x', labelbottom=True, labelsize=8)

axes[0].tick_params(axis='x', labelbottom=True)
plt.tight_layout()
plt.xticks(rotation=60)
#fig.savefig(symbol + interval + period + '.png', dpi=300, bbox_inches="tight")

print(symbol)
print('Medium', pdf.index[pdf['Medium_signal']].to_list())
print('Strong', pdf.index[pdf['Strong_signal']].to_list())

plt.show()
```
