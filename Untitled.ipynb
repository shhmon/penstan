```python
"""
PENNY STOCK SPIKE DETECTION & ANALYSIS
===============================================================================

PURPOSE:
This script analyzes OHLCV data to detect price spikes in penny/low-cap stocks.
Primary goal: Detect WHEN a spike is happening (0-2 bars in), not predict days ahead.
Data source: yfinance (may have missing volume in pre/post market)

KEY DESIGN DECISIONS:
1. INCLUDES pre/post market data (critical for penny stocks)
2. Volume smoothing (2-bar rolling) to reduce yfinance noise
3. Volume metrics marked as NaN during pre/post when unreliable
4. Primary signals focus on PRICE ACTION (gaps, range, HOD breaks)
5. Volume used as CONFIRMATION when available, not requirement

SIGNAL PRIORITY (from highest to lowest confidence):
-  Spike_detected: Main signal - catches spikes as they happen
-  Price_breakout: Pure price action, works in pre/post
-  Gap_and_run: Gap up with continuation
-  Volatility_breakout: ATR expansion with momentum
-  Momentum_building: Early warning, lower threshold

USAGE:
Typical workflow is:
1. Social media/scanner provides tickers
2. Run this analysis to detect spike timing
3. Signals show WHERE in the chart the action is happening
4. Use for entry timing, not for finding tickers

VOLUME HANDLING:
- yfinance sets volume=0 for pre/post market bars
- We mark these with Is_prepost flag
- Volume metrics use NaN instead of 0 to avoid false spikes
- Volume confirmation skipped for pre/post bars
- Gray shading on charts shows pre/post periods

===============================================================================
"""

# Suppress performance warnings (we handle them)
warnings.filterwarnings('ignore', category=pd.errors.PerformanceWarning)

def make_gains(df, symbol, save=False, std_win=8):
    # Defragment DataFrame at the start to avoid performance issues
    pdf = df.copy()
    
    # Need to convert dates to string to easily align bar charts and lines
    pdf.index = pdf.index.astype(str)
    
    # =============================================================================
    # DATA PREPARATION & VOLUME HANDLING
    # =============================================================================
    
    # Light smoothing to reduce yfinance noise (apply before other processing)
    pdf['Volume_raw'] = pdf['Volume'].copy()
    pdf['Volume'] = pdf['Volume'].rolling(window=2).mean()
    
    # Mark regular vs pre/post market hours (yfinance sets volume=0 for pre/post)
    pdf['Is_regular_hours'] = pdf['Volume_raw'] > 0
    pdf['Is_prepost'] = ~pdf['Is_regular_hours']
    
    # Create volume metrics that handle missing pre/post data
    pdf['Volume_valid'] = pdf['Volume'].replace(0, np.nan)
    pdf['Volume_ewm_base'] = pdf['Volume_valid'].ewm(span=std_win, adjust=False).mean()
    pdf['Volume_ewm'] = pdf['Volume_ewm_base'].ffill()  # Forward fill for continuity
    
    # =============================================================================
    # BASIC METRICS
    # =============================================================================
    
    # --- Normalization ---

    
    # --- Volume Metrics (with pre/post handling) ---
    # Volume ratio: only meaningful during regular hours
    pdf['Volume_ratio'] = np.where(
        pdf['Is_regular_hours'],
        pdf['Volume'] / pdf['Volume_ewm'],
        np.nan
    )
    pdf['Volume_pct_change'] = pdf['Volume'].pct_change().fillna(0)
    pdf['Volume_pct_change_regular'] = np.where(
        pdf['Is_regular_hours'],
        pdf['Volume_pct_change'],
        np.nan
    )
    
    # --- Price Metrics ---
    pdf['Close_pct_change'] = pdf['Close'].pct_change().fillna(0)
    pdf['Close_roll'] = pdf['Close'].rolling(window=std_win).mean()
    pdf['Close_roll_pct_change'] = pdf['Close_roll'].pct_change().fillna(0)
    
    # --- Candle Metrics ---
    pdf['Gain'] = pdf['Close'] - pdf['Open']
    pdf['Range'] = pdf['High'] - pdf['Low']
    pdf['Gain_close_ratio'] = pdf['Gain'] / pdf['Close']
    
    # --- Support/Resistance ---
    pdf['Resistance'] = pdf['High'].rolling(window=std_win*2).max()
    pdf['Support'] = pdf['Low'].rolling(window=std_win*2).min()
    pdf['Distance_to_resistance'] = (pdf['Resistance'] - pdf['Close']) / pdf['Close']
    pdf['Distance_to_support'] = (pdf['Close'] - pdf['Support']) / pdf['Close']
    
    # --- Volatility Metrics ---
    pdf['TR'] = true_range(pdf)
    pdf['ATR'] = pdf['TR'].rolling(window=std_win*2).mean()
    pdf['ATR_pct'] = pdf['ATR'] / pdf['Close']  # Normalized ATR
    pdf['ATR_trend'] = pdf['ATR'].diff()
    pdf['ATR_compression'] = (pdf['ATR'] < pdf['ATR'].rolling(std_win*2).mean() * 0.8)
    
    # --- Bollinger Bands ---
    (pdf['Boll_h'], pdf['Boll_m'], pdf['Boll_l']) = bollinger_bands(pdf['Close'], length=std_win*2)
    pdf['Boll_w'] = pdf['Boll_h'] - pdf['Boll_l']
    pdf['Boll_w_avg'] = pdf['Boll_w'].rolling(std_win*2).mean()
    pdf['Boll_squeeze'] = (pdf['Boll_w'] < pdf['Boll_w_avg'] * 0.7)
    pdf['Boll_breakout_h'] = (pdf['Close'] > pdf['Boll_h'])
    pdf['Boll_breakout_l'] = (pdf['Close'] < pdf['Boll_l'])
    pdf['Bollinger_pct'] = (pdf['Close'] - pdf['Boll_l']) / (pdf['Boll_h'] - pdf['Boll_l'])
    
    # --- RSI ---
    pdf['RSI'] = calculate_rsi_ema(pdf['Close'], std_win*2)
    pdf['RSI_fast'] = calculate_rsi_ema(pdf['Close'], std_win)
    pdf['RSI_slow'] = calculate_rsi_ema(pdf['Close'], std_win*3)
    
    # =============================================================================
    # ADVANCED VOLUME METRICS (with pre/post awareness)
    # =============================================================================
    
    # --- Volume Acceleration & Persistence ---
    pdf['Volume_acceleration'] = pdf['Volume_ratio'].diff()
    pdf['Volume_elevated_count'] = (pdf['Volume_ratio'] > 2.0).rolling(std_win).sum()
    
    # --- Volume Trend (is volume increasing?) ---
    def calc_volume_trend(series, window):
        """Returns 1 if volume is trending up, 0 otherwise"""
        def trend_check(x):
            if len(x) < 2 or x.isna().any():
                return 0
            slope = np.polyfit(range(len(x)), x, 1)[0]
            return 1 if slope > 0 else 0
        return series.rolling(window).apply(trend_check, raw=False)
    
    pdf['Volume_trend'] = calc_volume_trend(pdf['Volume_ratio'], std_win)
    
    # --- Multi-timeframe Volume Ratios ---
    pdf['Volume_ratio_fast'] = np.where(
        pdf['Is_regular_hours'],
        pdf['Volume'] / pdf['Volume'].ewm(span=std_win//2).mean(),
        np.nan
    )
    pdf['Volume_ratio_slow'] = np.where(
        pdf['Is_regular_hours'],
        pdf['Volume'] / pdf['Volume'].ewm(span=std_win*2).mean(),
        np.nan
    )
    
    # --- Volume Momentum (z-score) ---
    rolling_mean = pdf['Volume'].rolling(std_win).mean()
    rolling_std = pdf['Volume'].rolling(std_win).std()
    pdf['Volume_momentum_norm'] = (pdf['Volume'] - rolling_mean) / rolling_std.replace(0, np.nan)
    
    # Mark when volume is available for confirmation
    pdf['Volume_confirmable'] = pdf['Is_regular_hours'] & (pdf['Volume_ratio'].notna())
    
    # =============================================================================
    # ADVANCED PRICE ACTION METRICS
    # =============================================================================
    
    # --- Candle Body & Shadow Analysis ---
    pdf['Body_range_ratio'] = abs(pdf['Gain']) / pdf['Range'].replace(0, np.nan)
    pdf['Upper_shadow'] = pdf['High'] - pdf[['Open', 'Close']].max(axis=1)
    pdf['Lower_shadow'] = pdf[['Open', 'Close']].min(axis=1) - pdf['Low']
    pdf['Shadow_ratio'] = pdf['Lower_shadow'] / (pdf['Upper_shadow'].replace(0, 0.0001))
    
    # --- Price Momentum Multi-timeframe ---
    pdf['Close_roc_fast'] = pdf['Close'].pct_change(periods=std_win//2)
    pdf['Close_roc_slow'] = pdf['Close'].pct_change(periods=std_win*2)
    
    # --- Price Efficiency (trending vs choppy) ---
    def efficiency_ratio(series, period):
        """Measures how efficiently price moves (1=straight line, 0=choppy)"""
        change = abs(series.diff(period))
        volatility = series.diff().abs().rolling(period).sum()
        return change / volatility.replace(0, np.nan)
    
    pdf['Price_efficiency'] = efficiency_ratio(pdf['Close'], std_win)
    
    # --- Range Metrics ---
    pdf['Range_pct'] = pdf['Range'] / pdf['Close']
    pdf['Range_expansion'] = pdf['Range_pct'] > pdf['Range_pct'].rolling(std_win).mean() * 1.5
    
    # --- Gap Analysis (CRITICAL for penny stocks) ---
    pdf['Gap'] = pdf['Open'] - pdf['Close'].shift(1)
    pdf['Gap_pct'] = pdf['Gap'] / pdf['Close'].shift(1)
    pdf['Gap_up'] = pdf['Gap_pct'] > 0.01  # 1% gap up
    pdf['Big_gap_up'] = pdf['Gap_pct'] > 0.10  # 10%+ gap (captures pre/post moves)
    pdf['Big_gap_down'] = pdf['Gap_pct'] < -0.10
    
    # --- Explosive Range (key for penny stocks) ---
    pdf['Range_pct'] = pdf['Range'] / pdf['Open']  # Range as % of open
    pdf['Explosive_range'] = pdf['Range_pct'] > 0.20  # 20%+ intraday range
    
    # --- Price Velocity (rate of change) ---
    pdf['Price_velocity'] = pdf['Close'].pct_change(periods=1)
    pdf['Rapid_price_move'] = abs(pdf['Price_velocity']) > 0.05  # 5% per bar
    
    # --- High of Day Breaks ---
    pdf['HOD'] = pdf['High'].expanding().max()  # High of day
    pdf['New_HOD'] = pdf['High'] >= pdf['HOD'].shift(1)  # Breaking into new HOD
    pdf['HOD_break_confirmed'] = pdf['New_HOD'] & (pdf['Close'] > pdf['Open'])  # Green at HOD
    
    # --- Consecutive Green Candles & Momentum ---
    pdf['Green_streak'] = (pdf['Gain'] > 0).astype(int).groupby(
        (pdf['Gain'] <= 0).astype(int).cumsum()
    ).cumsum()
    pdf['Spike_momentum'] = pdf['Green_streak'] >= 3  # 3+ green bars = momentum
    
    # =============================================================================
    # VOLUME-PRICE RELATIONSHIP METRICS (adjusted for pre/post)
    # =============================================================================
    
    # --- Volume Sentiment (Buying/Selling Pressure within candle) ---
    pdf['Volume_sentiment'] = np.where(
        (pdf['High'] != pdf['Low']) & pdf['Is_regular_hours'],
        ((pdf['Close'] - pdf['Low']) - (pdf['High'] - pdf['Close'])) / 
        (pdf['High'] - pdf['Low']) * pdf['Volume'],
        0
    )
    
    # --- On-Balance Volume (OBV) - only use when volume is valid ---
    pdf['OBV'] = (np.sign(pdf['Close'].diff()) * pdf['Volume_valid']).fillna(0).cumsum()
    pdf['OBV_change'] = pdf['OBV'].pct_change(periods=std_win).fillna(0)
    
    # --- Money Flow ---
    pdf['Typical_price'] = (pdf['High'] + pdf['Low'] + pdf['Close']) / 3
    pdf['Money_flow'] = pdf['Typical_price'] * pdf['Volume_valid'].fillna(0)
    pdf['Money_flow_ratio'] = pdf['Money_flow'] / pdf['Money_flow'].rolling(std_win).mean()
    
    # --- Buying vs Selling Pressure ---
    pdf['Buying_pressure'] = np.where(
        pdf['Is_regular_hours'] & (pdf['Range'] > 0),  # Only when range exists
        pdf['Volume'] * ((pdf['Close'] - pdf['Low']) / pdf['Range']),
        0  # Use 0 instead of NaN for aggregation
    )
    pdf['Selling_pressure'] = np.where(
        pdf['Is_regular_hours'] & (pdf['Range'] > 0),  # Only when range exists
        pdf['Volume'] * ((pdf['High'] - pdf['Close']) / pdf['Range']),
        0  # Use 0 instead of NaN for aggregation
    )
    
    # Calculate ratio with safe division
    buying_sum = pdf['Buying_pressure'].rolling(std_win).sum()
    selling_sum = pdf['Selling_pressure'].rolling(std_win).sum()
    pdf['Buy_sell_ratio'] = np.where(
        selling_sum > 0,
        buying_sum / selling_sum,
        np.nan
    )
    
    # =============================================================================
    # DIVERGENCE DETECTION
    # =============================================================================
    
    def detect_bullish_divergence(price, indicator, window):
        """Detects bullish divergence: price down, indicator up"""
        price_trend = price.diff(window)
        ind_trend = indicator.diff(window)
        return (price_trend < 0) & (ind_trend > 0)
    
    pdf['Bullish_divergence_OBV'] = detect_bullish_divergence(pdf['Close'], pdf['OBV'], std_win)
    pdf['Bullish_divergence_RSI'] = detect_bullish_divergence(pdf['Close'], pdf['RSI'], std_win)
    
    # =============================================================================
    # SIMPLE SIGNAL COMPONENTS
    # =============================================================================
    
    # --- Volume Signals (only during regular hours) ---
    pdf['Volume_spike'] = (
        pdf['Is_regular_hours'] & 
        (pdf['Volume_pct_change_regular'] > 5.0)  # 500% increase
    )
    pdf['Volume_spike_near'] = (pdf['Volume_spike'].astype(int).rolling(std_win).sum() > 0)
    pdf['Momentum_spike'] = (
        pdf['Volume_confirmable'] & 
        (pdf['Volume_momentum_norm'] > 2.0)  # 2 std devs
    )
    
    # --- Sentiment Signals ---
    rolling_sentiment = pdf['Volume_sentiment'].rolling(std_win*2)
    pdf['Bullish_sentiment'] = (
        pdf['Volume_sentiment'] > (rolling_sentiment.mean() + 2 * rolling_sentiment.std())
    )
    
    # --- Bollinger Signals ---
    pdf['Boll_squeeze_near'] = (pdf['Boll_squeeze'].astype(int).rolling(std_win).sum() > 0)
    
    # --- Volatility Signals ---
    pdf['Volatility_expansion'] = (pdf['ATR'] > pdf['ATR'].rolling(std_win).mean() * 1.5)
    
    # --- Price Signals ---
    pdf['Sustained_gain'] = (
        (pdf['Gain'] > 0) & 
        (pdf['Gain'].rolling(std_win//2).apply(lambda x: (x > 0).sum()) >= std_win//2 * 0.75)
    )
    
    # --- RSI Signals ---
    pdf['RSI_pos'] = (pdf['RSI'] >= 60)
    pdf['RSI_neg'] = (pdf['RSI'] <= 40)
    pdf['RSI_aligned_bullish'] = (pdf['RSI_fast'] > 50) & (pdf['RSI_slow'] > 50)
    
    # --- Breakout Signal ---
    pdf['Break'] = (abs(pdf['Gain']) > 1.9 * pdf['ATR']) | (pdf['Range'] > 1.9 * pdf['ATR'])
    
    # =============================================================================
    # PENNY STOCK SPIKE DETECTION SIGNALS
    # =============================================================================
    
    # --- SPIKE DETECTED: Early detection (0-2 bars into spike) ---
    # Primary signal for detecting when a spike is happening NOW
    pdf['Spike_detected'] = (
        (
            # Price-based (always reliable, no volume dependency)
            (pdf['Big_gap_up'] | pdf['Explosive_range']) &    # Unusual price movement
            (pdf['Range_pct'] > 0.15) &                        # At least 15% range
            (pdf['Close'] > pdf['Open']) &                     # Green candle
            (pdf['Body_range_ratio'] > 0.5) &                  # Decent body (not all wicks)
            (
                pdf['New_HOD'] |                               # Breaking high of day OR
                (pdf['Close'] > pdf['Boll_h'])                 # Breaking Bollinger upper
            )
        ) &
        (
            # Volume confirmation (when available, otherwise skip)
            (
                pdf['Is_prepost'] |                            # Skip volume check in pre/post OR
                (pdf['Volume_ratio'] > 2.0)                    # High volume during regular
            )
        ) &
        (
            # Momentum confirmation
            (pdf['Spike_momentum'] | pdf['Rapid_price_move'])  # Multiple greens OR rapid move
        )
    )
    
    # --- PRICE ACTION BREAKOUT: Pure price-based (no volume) ---
    # Works in pre/post market when volume is missing
    pdf['Price_breakout'] = (
        (pdf['Close'] > pdf['Boll_h']) &                       # Above upper band
        (pdf['Explosive_range'] | pdf['Big_gap_up']) &         # Large move
        (pdf['Close'] > pdf['Open']) &                         # Green
        (pdf['Body_range_ratio'] > 0.6) &                      # Strong body
        (pdf['New_HOD'])                                       # New high of day
    )
    
    # --- VOLATILITY SPIKE: Range expansion ---
    pdf['Volatility_breakout'] = (
        (pdf['ATR'] > pdf['ATR'].rolling(std_win).mean() * 1.5) &  # ATR expanding
        (pdf['Range_expansion']) &                                  # Range > average
        (pdf['Rapid_price_move']) &                                 # Fast price move
        (pdf['Close'] > pdf['Open'])                                # Green
    )
    
    # --- MOMENTUM BUILDING: Early signs before full spike ---
    # Lower threshold, catches more but with less certainty
    pdf['Momentum_building'] = (
        (pdf['Range_pct'] > 0.08) &                            # 8%+ range (lower threshold)
        (pdf['Green_streak'] >= 2) &                           # 2+ consecutive greens
        (pdf['Close'] > pdf['Boll_m']) &                       # Above mid-band
        (
            pdf['Is_prepost'] |                                # Skip volume in pre/post OR
            (pdf['Volume_ratio'] > 1.5)                        # Above average volume
        )
    )
    
    # --- GAP AND RUN: Gap up followed by continuation ---
    pdf['Gap_and_run'] = (
        (pdf['Big_gap_up']) &                                  # Gapped up significantly
        (pdf['Close'] > pdf['Open']) &                         # Continued upward (green)
        (pdf['New_HOD']) &                                     # Made new high
        (pdf['Body_range_ratio'] > 0.5)                        # Strong conviction
    )
    
    # =============================================================================
    # LEGACY COMPOSITE SIGNALS (now with pre/post handling)
    # =============================================================================
    
    # --- PRE-BREAKOUT: Compression phase before the move ---
    pdf['Pre_breakout'] = (
        (pdf['ATR_compression']) &                             # Volatility contracting
        (pdf['Boll_squeeze']) &                                # Price range tight
        (pdf['Volume_elevated_count'].fillna(0) >= 2) &        # Volume building (if available)
        (pdf['Volume_trend'].fillna(0) == 1) &                 # Volume trending up
        (pdf['Distance_to_resistance'] < 0.05) &               # Near resistance
        (pdf['Body_range_ratio'] > 0.5)                        # Strong conviction
    )
    
    # --- ACCUMULATION: Smart money building positions ---
    pdf['Accumulation'] = (
        (pdf['Price_efficiency'] < 0.3) &                      # Price choppy
        (pdf['OBV_change'] > 0.05) &                           # OBV rising
        (pdf['Buy_sell_ratio'].fillna(1) > 1.3) &              # More buying (if available)
        (pdf['Volume_ratio'].fillna(0) > 1.5) &                # Above average volume
        (pdf['Volume_elevated_count'].fillna(0) >= 3) &        # Sustained volume
        (~pdf['Boll_breakout_h']) &                            # Haven't broken out
        (pdf['Is_regular_hours'])                              # Only during regular hours
    )
    
    # --- MOMENTUM IGNITION: Early stage breakout ---
    pdf['Momentum_ignition'] = (
        (pdf['Volume_acceleration'].fillna(0) > 0) &           # Volume accelerating
        (pdf['Volume_ratio'].fillna(0) > 3) &                  # Strong volume (if available)
        (pdf['Close'] > pdf['Boll_m']) &                       # Above mid-band
        (pdf['Money_flow_ratio'].fillna(1) > 2) &              # Money flowing in
        (pdf['Green_streak'] >= 2) &                           # Consecutive greens
        (pdf['ATR_trend'] > 0) &                               # Volatility expanding
        (pdf['RSI'] > 45) & (pdf['RSI'] < 70)                  # RSI has room
    )
    
    # --- BREAKOUT CONFIRMED: For entering after move starts ---
    pdf['Breakout_confirmed'] = (
        (pdf['Pre_breakout'].shift(1) | pdf['Pre_breakout'].shift(2)) &  # Was coiling
        (pdf['Close'] > pdf['Boll_h']) &                       # Broke upper band
        (pdf['Volume_ratio'].fillna(0) > 4) &                  # Strong volume (if available)
        (pdf['Range_expansion']) &                             # Range expanding
        (pdf['Body_range_ratio'] > 0.6) &                      # Strong conviction
        (pdf['Shadow_ratio'] > 1)                              # Bullish structure
    )
    
    # --- LEGACY SIGNALS (from original code) ---
    pdf['Price_vol_breakout'] = (pdf['Close_roll_pct_change'] > 0.1) & (pdf['Volume_ratio'] > 3)
    
    pdf['Medium_signal'] = (
        (pdf['Boll_breakout_h']) &
        (pdf['Volume_spike_near'] | pdf['Volatility_expansion']) &
        (pdf['RSI_pos']) &
        (pdf['Gain'].diff() > 0)
    )
    
    pdf['Strong_signal'] = (
        (pdf['Medium_signal']) &
        (pdf['Price_vol_breakout']) & 
        (pdf['Break']) & 
        (pdf['Sustained_gain']) &
        (pdf['Volume_sentiment'] > 0) &
        (pdf['Bullish_sentiment'])
    )
    
    # =============================================================================
    # PLOTTING
    # =============================================================================
    
    fig, axes = plt.subplots(7, 2, figsize=(14, 20), sharex=True)
    axes = axes.flatten()
    
    def scatter_dot(ax, column, y, color=None, label=None):
        """Plot signal indicators as scatter points"""
        if column not in pdf.columns:
            return
        signal_positions = pdf.index[pdf[column].astype(bool)]
        if len(signal_positions) == 0:
            return
        signal_positions = [pdf.index.get_loc(dt) for dt in signal_positions]
        ax.scatter(signal_positions, [y]*len(signal_positions), 
                   color=color, s=12, zorder=6, label=label or column, marker='|', linewidths=2)
    
    # --- Plot 0: Main Chart with Signals ---
    ax = axes[0]
    midy = max(pdf['Boll_h'].max(), pdf['Close'].max()) / 2
    
    # NEW PENNY STOCK SIGNALS (Priority signals) - NO EMOJIS, BETTER COLORS
    scatter_dot(ax, 'Spike_detected', y=midy*2.15, color='darkred', label='SPIKE DETECTED')
    scatter_dot(ax, 'Price_breakout', y=midy*2.10, color='red', label='Price Breakout')
    scatter_dot(ax, 'Gap_and_run', y=midy*2.05, color='darkorange', label='Gap & Run')
    scatter_dot(ax, 'Volatility_breakout', y=midy*2.00, color='purple', label='Volatility Break')
    scatter_dot(ax, 'Momentum_building', y=midy*1.95, color='darkgoldenrod', label='Momentum Building')
    
    # Legacy signals (less emphasis)
    scatter_dot(ax, 'Pre_breakout', y=midy*1.90, color='cyan', label='Pre-breakout')
    scatter_dot(ax, 'Momentum_ignition', y=midy*1.85, color='pink', label='Momentum Ignition')
    
    # Component signals (bottom of chart)
    scatter_dot(ax, 'New_HOD', y=0.1*midy, color='lime', label='New HOD')
    scatter_dot(ax, 'Boll_breakout_h', y=0.2*midy, color='green', label='BB Break')
    scatter_dot(ax, 'Big_gap_up', y=0.3*midy, color='gold', label='Big Gap')
    scatter_dot(ax, 'Explosive_range', y=0.4*midy, color='crimson', label='Explosive Range')
    scatter_dot(ax, 'Volume_spike', y=0.5*midy, color='blue', label='Vol Spike')
    
    # Pre/post market hours shading
    for idx in pdf.index[pdf['Is_prepost']]:
        pos = pdf.index.get_loc(idx)
        ax.axvspan(pos-0.5, pos+0.5, alpha=0.05, color='gray', zorder=1)
    
    # Price and Bollinger Bands
    pdf.plot(y='Close', ax=ax, color='black', linewidth=1.5, label='Close')
    pdf.plot(y='Boll_h', ax=ax, alpha=0.6, color='skyblue', label='BB Upper')
    pdf.plot(y='Boll_m', ax=ax, alpha=0.4, color='gray', linestyle='--', linewidth=0.8)
    pdf.plot(y='Boll_l', ax=ax, alpha=0.6, color='skyblue', label='BB Lower')
    pdf.plot(y='Boll_w', kind='bar', ax=ax, alpha=0.1, width=1, color='slategray')
    pdf[pdf['Boll_squeeze']].plot(y='Boll_w', kind='bar', ax=ax, alpha=0.3, width=1, color='yellow')
    ax.set_title('Price, Bollinger Bands & Signals', fontweight='bold')
    ax.set_ylabel('Price')
    
    # --- Plot 1: Normalized Close & Volume ---
    ax = axes[1]
    pdf.plot(y='Volume_n', ax=ax, kind='bar', alpha=0.4, width=1, color='steelblue', label='Volume')
    pdf.plot(y='Close_n', ax=ax, kind='line', color='black', linewidth=1, label='Close')
    ax.set_title('Normalized Close & Volume')
    ax.set_ylabel('Normalized (0-1)')
    
    # --- Plot 2: ATR & Gain with Break Points ---
    ax = axes[2]
    break_points = pdf.index[pdf['Break'].astype(bool)]
    break_positions = [pdf.index.get_loc(dt) for dt in break_points]
    if len(break_positions) > 0:
        ax.scatter(break_positions, [0]*len(break_positions), 
                   color='red', s=20, zorder=5, label='Break', marker='x')
    pdf.plot(y='ATR', ax=ax, kind='bar', alpha=0.4, width=1, color='skyblue', label='ATR')
    pdf.plot(y='Gain', ax=ax, kind='bar', alpha=0.5, width=1, color='orange', label='Gain')
    ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    ax.set_title('ATR, Gain & Breakout Points')
    ax.set_ylabel('Value')
    
    # --- Plot 3: Volume Metrics ---
    ax = axes[3]
    pdf.plot(y='Volume_ratio', ax=ax, color='darkblue', linewidth=2, label='Volume Ratio')
    pdf.plot(y='Volume_ratio_fast', ax=ax, color='blue', alpha=0.7, linewidth=1.5, label='Vol Ratio Fast')
    pdf.plot(y='Volume_ratio_slow', ax=ax, color='navy', alpha=0.7, linewidth=1.5, label='Vol Ratio Slow')
    ax.axhline(y=1, color='gray', linestyle='--', linewidth=0.8)
    ax.axhline(y=2, color='orange', linestyle='--', linewidth=0.8, alpha=0.5)
    ax.axhline(y=3, color='red', linestyle='--', linewidth=0.8, alpha=0.5)
    ax.set_title('Volume Ratios (Multi-timeframe)')
    ax.set_ylabel('Ratio')
    
    # --- Plot 4: RSI ---
    ax = axes[4]
    pdf.plot(y='RSI', ax=ax, kind='line', color='indigo', linewidth=1.5, label='RSI')
    pdf.plot(y='RSI_fast', ax=ax, kind='line', color='blue', alpha=0.5, linewidth=1, label='RSI Fast')
    pdf.plot(y='RSI_slow', ax=ax, kind='line', color='purple', alpha=0.5, linewidth=1, label='RSI Slow')
    ax.axhline(y=70, color='red', linestyle='--', linewidth=1, label='Overbought')
    ax.axhline(y=30, color='green', linestyle='--', linewidth=1, label='Oversold')
    ax.axhline(y=50, color='gray', linestyle='-', linewidth=0.5, alpha=0.5)
    ax.set_ylim(0, 100)
    ax.set_title('RSI (Multi-timeframe)')
    ax.set_ylabel('RSI')
    
    # --- Plot 5: Buying/Selling Pressure ---
    ax = axes[5]
    pdf.plot(y='Buy_sell_ratio', ax=ax, color='green', linewidth=1.5, label='Buy/Sell Ratio')
    ax.axhline(y=1, color='gray', linestyle='--', linewidth=1, label='Equilibrium')
    ax.axhline(y=1.5, color='lightgreen', linestyle='--', linewidth=0.8, alpha=0.5)
    ax.fill_between(range(len(pdf)), 1, pdf['Buy_sell_ratio'], 
                     where=(pdf['Buy_sell_ratio'] >= 1), alpha=0.2, color='green', interpolate=True)
    ax.fill_between(range(len(pdf)), 1, pdf['Buy_sell_ratio'], 
                     where=(pdf['Buy_sell_ratio'] < 1), alpha=0.2, color='red', interpolate=True)
    ax.set_title('Buying vs Selling Pressure')
    ax.set_ylabel('Ratio')
    
    # --- Plot 6: On-Balance Volume ---
    ax = axes[6]
    pdf.plot(y='OBV', ax=ax, color='purple', linewidth=1.5, label='OBV')
    ax.set_title('On-Balance Volume (OBV)')
    ax.set_ylabel('Cumulative Volume')
    ax2 = ax.twinx()
    pdf.plot(y='OBV_change', ax=ax2, kind='bar', alpha=0.3, width=1, color='violet', label='OBV Change %')
    ax2.set_ylabel('OBV % Change', color='violet')
    ax2.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    
    # --- Plot 7: Volume Momentum ---
    ax = axes[7]
    pdf.plot(y='Volume_momentum_norm', ax=ax, kind='bar', color='teal', width=1, label='Vol Momentum Z-score')
    ax.axhline(y=2, color='red', linestyle='--', linewidth=1, alpha=0.7, label='2')
    ax.axhline(y=-2, color='green', linestyle='--', linewidth=1, alpha=0.7, label='-2')
    ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    ax.set_title('Volume Momentum (Z-score)')
    ax.set_ylabel('Standard Deviations')
    
    # --- Plot 8: Money Flow ---
    ax = axes[8]
    pdf.plot(y='Money_flow_ratio', ax=ax, color='darkgreen', linewidth=1.5, label='Money Flow Ratio')
    ax.axhline(y=1, color='gray', linestyle='--', linewidth=1)
    ax.axhline(y=2, color='orange', linestyle='--', linewidth=0.8, alpha=0.5)
    ax.set_title('Money Flow Ratio')
    ax.set_ylabel('Ratio')
    
    # --- Plot 9: Volume Sentiment & Acceleration ---
    ax = axes[9]
    pdf.plot(y='Volume_sentiment', ax=ax, kind='bar', alpha=0.6, width=1, 
             color=pdf['Volume_sentiment'].apply(lambda x: 'green' if x > 0 else 'red'),
             label='Volume Sentiment')
    ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    ax2 = ax.twinx()
    pdf.plot(y='Volume_acceleration', ax=ax2, color='blue', linewidth=1, alpha=0.7, label='Vol Acceleration')
    ax2.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)
    ax.set_title('Volume Sentiment & Acceleration')
    ax.set_ylabel('Volume Sentiment')
    ax2.set_ylabel('Volume Acceleration', color='blue')
    
    # --- Plot 10: Price Efficiency & Body/Range Ratio ---
    ax = axes[10]
    pdf.plot(y='Price_efficiency', ax=ax, color='navy', linewidth=1.5, label='Price Efficiency')
    ax.axhline(y=0.3, color='orange', linestyle='--', linewidth=1, alpha=0.5, label='Choppy Threshold')
    ax2 = ax.twinx()
    pdf.plot(y='Body_range_ratio', ax=ax2, kind='bar', alpha=0.3, width=1, color='coral', label='Body/Range')
    ax.set_title('Price Efficiency & Candle Body Ratio')
    ax.set_ylabel('Price Efficiency', color='navy')
    ax2.set_ylabel('Body/Range Ratio', color='coral')
    
    # --- Plot 11: Gap Analysis & HOD Breaks ---
    ax = axes[11]
    pdf.plot(y='Gap_pct', ax=ax, kind='bar', width=1, 
             color=pdf['Gap_pct'].apply(lambda x: 'green' if x > 0 else 'red'),
             alpha=0.6, label='Gap %')
    ax.axhline(y=0.10, color='green', linestyle='--', linewidth=1, label='10% threshold')
    ax.axhline(y=-0.10, color='red', linestyle='--', linewidth=1)
    ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    scatter_dot(ax, 'New_HOD', y=pdf['Gap_pct'].max() * 0.8, color='gold', label='New HOD')
    ax.set_title('Gap Analysis & High of Day Breaks')
    ax.set_ylabel('Gap %')
    
    # Mark pre/post periods
    for idx in pdf.index[pdf['Is_prepost']]:
        pos = pdf.index.get_loc(idx)
        ax.axvspan(pos-0.5, pos+0.5, alpha=0.05, color='gray', zorder=1)
    
    # --- Plot 12: Distance to Support/Resistance ---
    ax = axes[12]
    pdf.plot(y='Distance_to_resistance', ax=ax, color='red', linewidth=1.5, label='Distance to Resistance')
    pdf.plot(y='Distance_to_support', ax=ax, color='green', linewidth=1.5, label='Distance to Support')
    ax.axhline(y=0.05, color='red', linestyle='--', linewidth=0.8, alpha=0.5, label='5% threshold')
    ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    ax.set_title('Distance to Support/Resistance')
    ax.set_ylabel('Distance (%)')
    
    # --- Plot 13: Volume with Pre/Post Marking ---
    ax = axes[13]
    # Color ALL bars based on regular vs pre/post - don't filter anything
    colors = ['orange' if is_reg else 'lightgray' for is_reg in pdf['Is_regular_hours']]
    ax.bar(range(len(pdf)), pdf['Volume'], width=1, color=colors, alpha=0.8, label='Volume')
    ax.set_yscale('log')
    ax.set_title('Volume (Orange=Regular, Gray=Pre/Post)')
    ax.set_ylabel('Volume (log scale)')
    ax.set_xlim(-0.5, len(pdf)-0.5)
    
    # Add volume ratio on secondary axis - show it for all bars but only meaningful during regular hours
    ax2 = ax.twinx()
    ax2.plot(range(len(pdf)), pdf['Volume_ratio'], color='darkblue', linewidth=2, alpha=0.8, label='Vol Ratio')
    ax2.set_ylabel('Volume Ratio', color='darkblue')
    ax2.axhline(y=2, color='red', linestyle='--', linewidth=0.8, alpha=0.5, label='2x')
    ax2.axhline(y=3, color='darkred', linestyle='--', linewidth=0.8, alpha=0.5, label='3x')
    ax2.legend(fontsize=7, loc='upper right')
    
    # Set x-axis
    ax.set_xticks(range(len(pdf)))
    ax.set_xticklabels([])  # Will be set by the common formatter below
    
    # =============================================================================
    # FORMAT ALL PLOTS
    # =============================================================================
    
    for ax in axes:
        # X-axis formatting
        xticks = ax.get_xticks()
        if len(xticks) > 25:
            ax.set_xticks(xticks[::2])
        
        labels = []
        for i in ax.get_xticks():
            if 0 <= int(i) < len(pdf):
                dt = pd.to_datetime(pdf.index[int(i)])
                labels.append(dt.strftime('%b %-d %H:%M'))
            else:
                labels.append('')
        ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=7)
        
        # Legend
        handles, labels = ax.get_legend_handles_labels()
        if handles:
            ax.legend(handles, labels, fontsize=7, loc='best')
        
        # Grid
        ax.grid(True, alpha=0.2, linestyle='--', linewidth=0.5)
        ax.tick_params(axis='both', labelsize=8)
    
    plt.tight_layout()
    
    # =============================================================================
    # PRINT SIGNAL SUMMARY
    # =============================================================================
    
    print(f"\n{'='*80}")
    print(f"SIGNAL SUMMARY FOR: {symbol}")
    print(f"{'='*80}\n")
    
    print(" DATA INFO:")
    print(f"  Total bars: {len(pdf)}")
    print(f"  Regular hours: {pdf['Is_regular_hours'].sum()}")
    print(f"  Pre/Post market: {pdf['Is_prepost'].sum()}")
    print(f"  Date range: {pdf.index[0]} to {pdf.index[-1]}")
    print()
    
    print(" PENNY STOCK SIGNALS (Primary):")
    print("-" * 80)
    signals_to_check = [
        (' SPIKE DETECTED', 'Spike_detected'),
        (' Price Breakout (no vol needed)', 'Price_breakout'),
        (' Gap and Run', 'Gap_and_run'),
        (' Volatility Breakout', 'Volatility_breakout'),
        (' Momentum Building', 'Momentum_building'),
    ]
    
    for label, col in signals_to_check:
        timestamps = pdf.index[pdf[col]].to_list()
        count = len(timestamps)
        print(f"{label:35} : {count:3} occurrences")
        if count > 0 and count <= 10:
            for ts in timestamps:
                dt = pd.to_datetime(ts)
                is_prepost = pdf.loc[ts, 'Is_prepost']
                marker = "  [PRE/POST]" if is_prepost else "  [REGULAR]"
                print(f"   {dt.strftime('%Y-%m-%d %H:%M')}{marker}")
        elif count > 10:
            print(f"   {timestamps[0]} ... {timestamps[-1]} (showing first and last)")
    
    print()
    print(" LEGACY SIGNALS:")
    print("-" * 80)
    legacy_signals = [
        ('Pre-Breakout', 'Pre_breakout'),
        ('Accumulation', 'Accumulation'),
        ('Momentum Ignition', 'Momentum_ignition'),
        ('Breakout Confirmed', 'Breakout_confirmed'),
    ]
    
    for label, col in legacy_signals:
        count = len(pdf.index[pdf[col]])
        print(f"{label:35} : {count:3} occurrences")
    
    print(f"\n{'='*80}\n")
    
    # =============================================================================
    # SHOW PLOT
    # =============================================================================
    
    plt.show()
    
    # Optional: Save figure
    if save:
        fig.savefig(f'{symbol}.png', dpi=300, bbox_inches='tight')
```

```python
# os.makedirs("figures", exist_ok=True)
import numpy as np
import matplotlib.dates as mdates
# MACD settings
short_window = 12
long_window = 26
signal_window = 9

# MSAI CMBM DFLI BYND ASST GPUS DVLT CSAI DTCK RLMD
# CHR FEMY NVVE EGHT NFE ETHZ QH ACRS CCLD IINN WORX OSRH

#Load data

df["date"] = df.index
df['date'] = pd.to_datetime(df['date'])

#EMAs and MACD
df['EMA12'] = df['Close'].ewm(span=short_window, adjust=False).mean()
df['EMA26'] = df['Close'].ewm(span=long_window, adjust=False).mean()
df['MACD'] = df['EMA12'] - df['EMA26']
df['Signal'] = df['MACD'].ewm(span=signal_window, adjust=False).mean()
df['hist_difference'] = df['MACD'] - df['Signal']
df['EMA200'] = df['Close'].ewm(span=200, adjust=False).mean()
df['trendDirection'] = np.where(df['Close'] > df['EMA200'], "up", "down")

#Buy Signal
df['Buy_Signal'] = 0
for i in range(1, len(df)):
    if (
        df['hist_difference'].iloc[i-1] < 0 and df['hist_difference'].iloc[i] > 0 and
        df['MACD'].iloc[i] < 0 and df['Signal'].iloc[i] < 0 and
        df['trendDirection'].iloc[i] == "up"
    ):
        df.loc[df.index[i], 'Buy_Signal'] = 1

#Portfolio Simulation
initial_cash = 1000
cash = initial_cash
in_trade = False
entry_price = 0
entry_date = None
trades = []
portfolio_values = []

stop_loss_pct = 0.02
profit_ratio = 1.5
take_profit_pct = stop_loss_pct * profit_ratio

for i in range(len(df)):
    price = df['Close'].iloc[i]
    date = df['date'].iloc[i]

    if not in_trade:
        if df['Buy_Signal'].iloc[i] == 1:
            entry_price = price
            stop_price = entry_price * (1 - stop_loss_pct)
            target_price = entry_price * (1 + take_profit_pct)
            entry_date = date
            in_trade = True
            print(f"BUY at {entry_price:.2f} on {date.date()} | Target: {target_price:.2f}, Stop: {stop_price:.2f}")
    else:
        if price >= target_price:
            cash = cash * (target_price / entry_price)
            print(f"SELL (TP HIT) at {target_price:.2f} on {date.date()} | Portfolio: {cash:.2f}")
            trades.append((entry_date, date, entry_price, target_price, 'win'))
            in_trade = False
        elif price <= stop_price:
            cash = cash * (stop_price / entry_price)
            print(f"SELL (STOP HIT) at {stop_price:.2f} on {date.date()} | Portfolio: {cash:.2f}")
            trades.append((entry_date, date, entry_price, stop_price, 'loss'))
            in_trade = False

    portfolio_values.append(cash)

df['Portfolio'] = portfolio_values

#figure 1: MACD, Histogram, and Price
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True,
                                    gridspec_kw={'height_ratios': [2, 1, 1]},
                                    figsize=(16, 9))

ax1.plot(df["date"], df["Close"], color="#00BFFF", linewidth=1.5, label="Close Price")
ax1.plot(df["date"], df["EMA200"], color="#FF5555", linestyle="--", linewidth=1.2, label="EMA 200")

buy_points = df[df['Buy_Signal'] == 1]
ax1.scatter(buy_points["date"], buy_points["Close"], color="#00FF88", marker="^", s=100, label="Buy Signal", zorder=5)

for trade in trades:
    start_date, end_date, entry, exit_price, result = trade
    color = "#00FF88" if result == "win" else "#FF5555"
    ax1.hlines(y=exit_price, xmin=start_date, xmax=end_date, colors=color, linestyles="--", linewidth=1.8)

ax1.legend(loc="upper left", fontsize=10)
ax1.set_ylabel("Price ($)")
ax1.grid(True, linestyle="--", alpha=0.3)

ax2.plot(df["date"], df["MACD"], color="#1E90FF", linewidth=1.2, label="MACD")
ax2.plot(df["date"], df["Signal"], color="#FFB347", linewidth=1.2, label="Signal")
ax2.axhline(0, color="#888888", linestyle="--", linewidth=1)
for _, row in buy_points.iterrows():
    ax2.annotate('', (row['date'], row['MACD']),
                 color='#00FF88', fontsize=12, ha='center')

ax2.legend(loc="upper left", fontsize=9)
ax2.set_ylabel("MACD")
ax2.grid(True, linestyle="--", alpha=0.3)

colors = ["#00FF88" if h >= 0 else "#FF5555" for h in df["hist_difference"]]
ax3.bar(df["date"], df["hist_difference"], color=colors, width=1)
ax3.axhline(0, color="#666666", linestyle="--", linewidth=1)
ax3.set_ylabel("Histogram")
ax3.set_xlabel("Date")
ax3.grid(True, linestyle="--", alpha=0.3)

ax3.xaxis.set_major_locator(mdates.YearLocator())
ax3.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
fig.autofmt_xdate()
plt.tight_layout()

# Save the first figure
fig.savefig("figures/figure1_trading_strategy.png", dpi=200, bbox_inches='tight')
plt.show()

#figure 2: Portfolio Value
fig2, ax4 = plt.subplots(figsize=(16, 9))
ax4.plot(df["date"], df["Portfolio"], color="#9370DB", linewidth=2.5, label="Portfolio Value")
ax4.set_title("Portfolio Value Over Time", color="white", fontsize=16)
ax4.set_xlabel("Date", color="white")
ax4.set_ylabel("Portfolio Value ($)", color="white")
ax4.grid(True, linestyle="--", alpha=0.3)
ax4.legend(facecolor="#111111", edgecolor="#222222", fontsize=10, loc="upper left")
ax4.xaxis.set_major_locator(mdates.YearLocator())
ax4.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
fig2.autofmt_xdate()
plt.tight_layout()

# Save the second figure
fig2.savefig("figures/figure2_portfolio_value.png", dpi=200, bbox_inches='tight')
plt.show()

#Accuracy scoring
if len(trades) > 0:
    total_trades = len(trades)
    wins = sum(1 for t in trades if t[4] == 'win')
    accuracy = (wins / total_trades) * 100
    print(f'Accuracy: {accuracy:.2f}% trades were profitable')
```
